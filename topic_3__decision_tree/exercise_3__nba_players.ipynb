```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```python
data = pd.read_csv('datasets/nba_2013.csv', sep=',')
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.shape
```

```python
# pick relevant columns for prediction
inputs = data.drop(['player','bref_team_id', 'season','season_end'], axis=1)
inputs.shape
```

```python
inputs.head()
```

```python
inputs = pd.get_dummies(inputs)
inputs.head()
```

```python
outputs = data['pts']
outputs = np.array(outputs)
outputs
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, outputs, test_size=0.3, random_state=42)
```

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
print('The Training accuracy score is: ', model.score(X_train, y_train)*100,'%')
print('The Testing accuracy score is: ', model.score(X_test, y_test)*100,'%')
```

```python
y_pred = model.predict(X_test)
print('MSE: ', metrics.mean_squared_error(y_test, y_pred))
```

> R^2 score of Train/Test set is similar and high at 99.57%-100%.
> MSE: 957
